##**Limpeza de Dados**
Neste tópico irei remover valores negativos das variáveis `new_cases` e `new_deaths`, pois, estes dados não condizem com a realidade, classificados como outliers abaixo do limite inferior.




```python
#criando um novo dataframe sem as entradas de continentes e do mundo

df_clean= df.loc[(df.iso_code != 'OWID_SAM') & (df.iso_code != 'OWID_AFR') & (df.iso_code != 'OWID_NAM') & (df.iso_code != 'OWID_OCE') & (df.iso_code != 'OWID_ASI') & (df.iso_code != 'OWID_EUR') & (df.iso_code != 'OWID_WRL') & (df.iso_code != 'OWID_EUN')].copy()
```

```python
#limpeza de valores negativos 

df_clean.drop(df_clean.loc[df_clean.new_cases < 0].index, axis=0, inplace=True)
df_clean.drop(df_clean.loc[df_clean.new_deaths < 0].index, axis=0, inplace=True)
```

Agora, irei verificar a porcentagem de valores ausentes nas variáveis `new_cases` e `total_cases`, pois elas serão usadas futuramente para novas análises.

```python
print("Porcentagem de valores ausentes na variável new_cases: {:.2f}%".format(df_clean.new_cases.isnull().sum() / df_clean.shape[0] * 100))
```

```python
print("Porcentagem de valores ausentes na variável total_cases: {:.2f}%".format(df_clean.total_cases.isnull().sum() / df_clean.shape[0] * 100))
```

Como podemos ver, os valores ausentes em ambas as variáveis são muito baixos, sendo assim, a remoção destas entradas será o melhor tratamento para esses valores ausentes.

```python
#removendo valores ausentes das variáveis new_cases e total_cases
df_clean.dropna(subset = ['new_cases'], axis= 0, inplace=True)
df_clean.dropna(subset = ['total_cases'], axis= 0, inplace=True)
```

```python
#verificando se realmente foram removidos os dados ausentes
df_clean.isnull().sum().sort_values(ascending=True)[:10]
```

**Muitas variáveis existem valores ausentes, mas nem todas irei utilizar em minha análise, por tanto não se faz necessário a remoção destes valores.**

Em contrapartida, existem países que contrariam novamente nossa ideia principal. Mas desta vez, **positivamente.**
* A Turquia, posicionada na 5ª posição dos países com os maiores números de novos casos, entretanto, ficou na 19ª posição dos países com mais mortes. Obtendo um "bom" resultado em relação novos casos versus óbitos.

Foi possível enxergar a regressão do Brasil em relação aos Estados Unidos desde o início de 2021, onde no país sul-americano teve um aumento de 178% no número de mortes e no país norte-americano teve uma diminuição em -78% no número de mortes.
